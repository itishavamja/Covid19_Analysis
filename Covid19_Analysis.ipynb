```python
# find maximum infection rate for all of the countries
countries = list(corona_dataset_aggregating.index)
max_infection_rates = []
for c in countries:
    # Convert the series to numeric before calculating the difference
    try:
        max_infection_rates.append(pd.to_numeric(corona_dataset_aggregating.loc[c]).diff().max())
    except ValueError:
        # Handle cases where conversion to numeric fails (e.g., non-numeric columns)
        print(f"Skipping country {c} due to non-numeric data.")
        max_infection_rates.append(np.nan)  # Or any other appropriate value
corona_dataset_aggregating["max_infection_rates"] = max_infection_rates
corona_dataset_aggregating.head(10)
```

```python
# create a new dataframe with only needed column
corona_data= pd.DataFrame(corona_dataset_aggregating["max_infection_rates"])
corona_data.head(10)
```

```python
# importing the worldHappinessReport.csv dataset
# selecting needed columns for our analysis
# join the dataset
# calculate the correlations as the result of our analysis


# importing the Dataset
happiness_report_csv=pd.read_csv("worldwide_happiness_report.csv")
```
